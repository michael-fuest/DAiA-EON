```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Preparation to read market data
path = './data/'
base = 'eex_power_base_'
peak = 'eex_power_peak_'
files = ["d", "w", "wkd", "q", "m", "y"]
gas = 'eex_gas_'
gas_files = ["d", "w", "wkd", "m", "q", "y", "s", "ss", "sw"]
spot_power = 'epex_spot_power'
spot_gas = 'eex_spot_gas'
csv = '.csv'
```

```python
power_merged = pd.read_csv(path+spot_power+csv, header = 0, parse_dates=[0], skiprows= lambda i: i>1932*24)
gas_merged = pd.read_csv(path+spot_gas+csv, header = 0, parse_dates=[0], skiprows= lambda i: i>1932*24)
```

```python
gas_merged.head()
```

```python
power_merged = power_merged.groupby(power_merged["time"].dt.date).mean().reset_index()
gas_merged = gas_merged.groupby(gas_merged["date"].dt.date).mean().reset_index()
```

```python
power_merged.columns = ["date", "spot"]
gas_merged.columns = ["date", "spot"]
```

```python
for file in files:
    pw_base = pd.read_csv(path+base+file+csv, header = 0, parse_dates = [0])
    pw_base["date"] = pw_base["date"].dt.date
    power_merged = power_merged.merge(pw_base, how="inner", on="date")
```

```python
for col in power_merged.columns:
    power_merged.rename(columns = {col : "base_"+col}, inplace = True)
```

```python
power_merged.rename(columns = {'base_date' : "date", 'base_spot' : "spot"}, inplace = True)
power_merged.head()
```

```python
peak_merged = pd.read_csv(path+spot_power+csv, header = 0, parse_dates=[0], skiprows= lambda i: i>1932*24)
peak_merged = peak_merged.groupby(peak_merged["time"].dt.date).mean().reset_index()
peak_merged.columns = ["date", "spot"]
```

```python
for file in files:
    pw_peak = pd.read_csv(path+peak+file+csv, header = 0, parse_dates = [0])
    pw_peak["date"] = pw_peak["date"].dt.date
    peak_merged = peak_merged.merge(pw_peak, how="inner", on="date")
```

```python
peak_merged.drop(columns = ["spot"], inplace = True)
```

```python
for col in peak_merged.columns:
    peak_merged.rename(columns = {col : 'peak_'+col}, inplace = True)
```

```python
peak_merged.rename(columns = {'peak_date' : "date"}, inplace = True)
peak_merged.head()
```

```python
power_merged = power_merged.merge(peak_merged, how ="inner", on="date")
power_merged.head()
```

```python
#power_merged.reset_index(drop=True, inplace=True)
power_merged.set_index('date', inplace=True)
```

```python
power_merged = power_merged.dropna(thresh=len(power_merged)*0.25, axis="columns")
#power_merged.interpolate(method='time', inplace=True)
power_merged.fillna(method="bfill", inplace=True)
power_merged.fillna(method="ffill", inplace=True)
```

```python
power_merged.isna().sum().sum()
```

```python
power_merged.head()
```

```python
# Defining periods for moving averages
periods = [7, 15, 30, 60, 100, 200]
```

```python
for col in power_merged.columns:
    for period in periods:
        power_merged[col+"_SMA"+str(period)] = power_merged[col].rolling(period).mean()
        power_merged[col+"_EWM"+str(period)] = power_merged[col].ewm(span=period).mean()
```

```python
power_merged.reset_index(inplace=True)
power_merged.tail()
```

```python
power_merged['date'] = pd.to_datetime(power_merged['date'])
power_merged['date'].dtypes
```

```python
date_range = (power_merged["date"]>='2021-05-25') & (power_merged["date"]<='2022-10-15')
```

```python
power_merged = power_merged.loc[date_range]
```

```python
power_merged.tail()
```

```python
power_merged.to_csv(path+"power_market_MA.csv")
```
