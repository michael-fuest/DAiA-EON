```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime, timedelta
```

```python
path='./data/'
store='./data/leaf_level/'
```

```python
supp_prices_power = pd.read_csv(path+'master_competitor_market_prices_power.csv')
```

```python
filename = []
for supplier in supp_prices_power['supplier'].unique():
    for postcode in supp_prices_power['post_code'].unique():
        for con_range in supp_prices_power['consumption_range_kwh'].unique():
            for tariff in supp_prices_power['tariff'].unique():
                df = supp_prices_power.loc[(supp_prices_power.supplier==supplier) & (supp_prices_power.post_code==postcode) &(supp_prices_power.consumption_range_kwh==con_range) &(supp_prices_power.tariff==tariff)].copy()
                if df.shape[0] > 0:
                    df["date"] = df.apply(lambda x: pd.date_range(x["date_valid_from"], x["date_valid_to"]), axis=1)
                    df = df.explode("date").copy()
                    print(df.head())
                    #df = df.groupby('date').agg({'price_year_eur':'mean'})
                    df.drop(columns = ["date_valid_from", "date_valid_to", "energy", "y_ahead_peak", "2y_ahead_peak", "3y_ahead_peak", "y_ahead_base", "2y_ahead_base", "3y_ahead_base", "d_ahead_base", "d_ahead_peak"], inplace=True)
                    df.reset_index(drop=True, inplace=True)
                else: 
                    continue    
                
                df.to_csv(store+str(supplier)+"_"+str(postcode)+"_"+str(con_range)+"_"+str(tariff)+".csv")
                new_file = str(str(supplier)+"_"+str(postcode)+"_"+str(con_range)+"_"+str(tariff)+".csv")
                filename.append(new_file)
                
```
