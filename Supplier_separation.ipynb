```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
```

```python
path='./data/'
store='./data/leaf_level/group/'
```

```python
supp_prices_power = pd.read_csv(path+'master_competitor_market_prices_power.csv', parse_dates=["date_valid_from", "date_valid_to"])
```

```python
supp_prices_power['price_per_kwh'] = supp_prices_power['price_year_eur']/supp_prices_power['consumption_range_kwh']
```

```python
supp_prices_power.head()
```

```python
supp_prices_power['supplier'].isna().sum()
```

```python
supp_prices_power['supplier'].nunique()
```

```python
filename = []
```

```python
count=1
for supplier in supp_prices_power['supplier'].unique():
    df=supp_prices_power.loc[supp_prices_power['supplier']==supplier].copy()
    if df.shape[0]>0:
        df["date"]=df.apply(lambda x: pd.date_range(x["date_valid_from"], x["date_valid_to"]), axis=1)
        df=df.explode("date").copy()
        df=df.groupby(['date']).agg({'price_per_kwh':'mean', 'strat_index':'mean'})
        df['Supplier_name'] = str(supplier)
        df.reset_index(inplace=True)
        print("Supplier no. ", count)
        count +=1
        df.to_csv(store+str(supplier)+".csv")
        new_file = str(str(supplier)+".csv")
        filename.append(new_file)
        
```
