```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
```

```python
path='./data/'
store='./data/leaf_level/group/'
```

```python
supp_prices_power = pd.read_csv(path+'master_competitor_market_prices_power.csv', parse_dates=["date_valid_from", "date_valid_to"])
```

```python
filename = []
```

```python
def write_to_csv(df, supplier, postcode, con_range, tariff):
    df1=df.copy()
    if df1.shape[0]>0:
        df1.to_csv(store+str(supplier)+"_"+str(postcode)+"_"+str(con_range)+"_"+str(tariff)+".csv")
        new_file = str(str(supplier)+"_"+str(postcode)+"_"+str(con_range)+"_"+str(tariff)+".csv")
        filename.append(new_file)
        
#supp_prices_power.groupby(['supplier', 'post_code', 'consumption_range_kwh', 'tariff']).apply(write_to_csv)
```

```python
def process_group(group, supplier, postcode, con_range, tariff):
    df=group.copy()
    if df.shape[0]>0:
        df["date"]=df.apply(lambda x: pd.date_range(x["date_valid_from"], x["date_valid_to"]), axis=1)
        df=df.explode("date").copy()
        df=df.groupby(['supplier', 'post_code', 'consumption_range_kwh', 'tariff', 'date']).agg({'price_year_eur':'mean', 'strat_index':'mode'})
        df.reset_index(inplace=True)
        write_to_csv(df, supplier, postcode, con_range, tariff)
    else:
        return pd.DataFrame()
    return df

supp_prices_power.groupby(['supplier', 'post_code', 'consumption_range_kwh', 'tariff']).apply(lambda x: process_group(x, x.name[0], x.name[1],x.name[2],x.name[3]))
    
```
